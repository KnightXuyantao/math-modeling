### 2.2 列表（最重要最重要最重要）

列表使用中括号[]

序列是 Python 中最基本的数据结构。

序列中的每个值都有对应的位置值，称之为索引，第一个索引是 0，第二个索引是 1，依此类推。

Python 有 6 个序列的内置类型，但最常见的是列表和元组。

列表都可以进行的操作包括索引，切片，加，乘，检查成员。

此外，Python 已经内置确定序列的长度以及确定最大和最小的元素的方法。

列表是最常用的 Python 数据类型，它可以作为一个方括号内的逗号分隔值出现。

列表的数据项不需要具有相同的类型

创建一个列表，只要把逗号分隔的不同的数据项使用方括号括起来即可。如下所示：

```python
list1 = ['Google', 'Runoob', 1997, 2000]
list2 = [1, 2, 3, 4, 5 ]
list3 = ["a", "b", "c", "d"]
list4 = ['red', 'green', 'blue', 'yellow', 'white', 'black']
print(list1, '\n', list2, '\n', list3, '\n', list4)
```

#### 访问列表中的值
与字符串的索引一样，列表索引从 0 开始，第二个索引是 1，依此类推。
通过索引列表可以进行截取、组合等操作。



```python
list = ['red', 'green', 'blue', 'yellow', 'white', 'black']
print( list[0] )
print( list[1] )
print( list[2] )
```

### 数组方法和常用函数

### 矩阵乘法

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("一维数组元素相乘", a*b) # 一维数组 元素相乘
print("------------------------------------------")
print("一维数组元素相乘再求和", np.dot(a, b)) # 一维数组元素相乘再求和
print("------------------------------------------")

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print("多维数组元素相乘\n", a*b) # 多维数组元素相乘
print("------------------------------------------")
print("多维数组实现矩阵相乘\n", np.dot(a, b)) # 多维数组实现矩阵相乘
print("------------------------------------------")

m = np.mat(a)
n = np.mat(b)

print("矩阵相乘1\n", np.dot(m, n)) # 矩阵相乘
print("------------------------------------------")
print("矩阵相乘2\n", m*n) # 矩阵相乘
```

### 1.3 线性代数模块

numpy.linalg 是numpy的线性代数模块，可以用来解决逆矩阵、特征值、线性方程组以及行列式等问题。

```python
m = np.mat('0 1 2; 1 0 3; 4 -3 8')
mi = np.linalg.inv(m) # 逆矩阵另一种方式
print(m)
print("------------------------------------------")
print("逆矩阵\n", mi)
print("------------------------------------------")
print(m*mi)
```

```python
m = np.mat('0 1 2; 1 0 3; 4 -3 8')
np.linalg.det(m) # 行列式
```

```python
m = np.mat('0 1 2; 1 0 3; 4 -3 8')
print("特征值", np.linalg.eigvals(m)) # 特征值
print("------------------------------------------")
print("特征向量\n", np.linalg.eig(m)) # 返回特征值及其对应特征向量的元组
```

### 求解线性方程组
$ x-2\cdot y+z=0 $

$2\cdot y-8\cdot z = 8$

$-4\cdot x+5\cdot y+9\cdot z=-9$

```python
A = np.mat('1 -2 1; 0 2 -8; -4 5 9')
b = np.array([0, 8, -9])
np.linalg.solve(A, b) # x = 29, y = 16, z = 3
```

## 3.Matplotlib(Pyplot)  画图用

```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定字体 KaiTi（楷体）
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像时‘-’显示为方块的问题
```

```python
x = np.arange(0, 2*np.pi, 0.01)
y = np.sin(x)
x_1 = np.arange(0, 2*np.pi, 0.1)
y_1 = np.cos(x_1)

plt.figure(figsize=(12, 8), dpi=100)
plt.plot(x,y, label='正弦曲线', c='r', lw=5) #
plt.scatter(x_1, y_1, label='余弦散点', marker='x', c='k') # 散点图
plt.title('图名')
plt.xlabel('横坐标名称')
plt.ylabel('纵坐标名称')
plt.legend()
#plt.savefig("123.png")
plt.show()
```

plt.plot函数的调用形式如下：

plot(x,y,color='green',linestyle='dashed',linewidth=1,marker='o',markerfacecolor='blue',markersize=6)

plot(x,y,c='g',ls='--',lw=1,marker='o',mfc='blue',ms=6)

color指定线的颜色，可简写为“c”。颜色的选项为：

蓝色： ‘b’ (blue)

绿色： ‘g’ (green)

红色： ‘r’ (red)

墨绿： ‘c’ (cyan)

洋红： ‘m’ (magenta)

黄色： ‘y’ (yellow)

黑色： ‘k’ (black)

白色： ‘w’ (white)

灰度表示： e.g. 0.75 ([0,1]内任意浮点数)

RGB表示法： e.g. ‘#2F4F4F’ 或 (0.18, 0.31, 0.31)

linestyle指定线型，可简写为“ls”。线型的选项为：

实线： ‘-’ (solid line)

### 使用numpy.polyfit拟合

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['KaiTi']
plt.rcParams['axes.unicode_minus'] = False

def f(t):
    """谷神星虚拟的运行轨道方程。我们假装不知道，仅用来验证预测结果是否准确"""
    
    t = t/7.5 -1
    return ((t**2-1)**3 + 0.5)*np.sin(2*t)

t = np.linspace(0, 20, 201) # 用于绘制实际的运行轨迹线

_x = np.linspace(0, 15, 16) # 观测数据时间序列
_y = f(_x) # 观测数据位置序列

x = np.linspace(15, 20, 6) # 待预测的时间序列

loss_list = []
for i in range(2,16): # 从2次到15次多项式，逐一计算误差
    args = np.polyfit(_x, _y, i) # 用最小二乘法找到最佳的一组系数
    g = np.poly1d(args) # 用这组系数生成方程g(x)
    loss = np.sum(np.square(g(_x)-_y)) # 计算i次多项式拟合的误差
    loss_list.append(loss)
    print(i, loss)

k = loss_list.index(min(loss_list))+2
args = np.polyfit(_x, _y, k)
g = np.poly1d(args)

plt.figure('demo', facecolor='#eaeaea', figsize=(12, 8), dpi=100)
plt.plot(_x, _y, c='r', ls='', marker='o', label=u'观测数据')
plt.plot(_x, g(_x), c='b', ls='-', label=u'%d次多项式拟合，误差%0.8f'%(k, loss_list[k-2]))
plt.plot(x, g(x), c='r', ls=':', label=u'预测轨迹')
plt.plot(t, f(t), c='#60f0f0', ls='--', label=u'实际运行轨迹')

plt.legend(loc='lower left')
plt.show()
print(g)
```

### 使用scipy.optimize.optimize.curve_fit拟合

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

x = np.arange(1, 13, 1)
y = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])

def fmax(x, a, b, c):
    """代拟合函数及参数a,b,c"""
    return a * np.sin(x * np.pi / 6 + b) + c

fita, fitb = optimize.curve_fit(fmax, x, y, [1, 1, 1])
print(fita) # 拟合系数 分别为上述函数的a, b, c

xn = np.arange(1, 13, 0.1)

plt.figure(figsize=(12, 8), dpi=100)
plt.plot(x, y)
plt.plot(xn, fmax(xn, fita[0],fita[1],fita[2]))
plt.show()
```

### 求解非线性方程(组)

在数学建模中，需要对一些稀奇古怪的方程（组）求解，Matlab自然是首选，但Matlab不是免费的，scipy则为我们提供了免费的午餐！scipy.optimize库中的fsolve函数可以用来对非线性方程（组）进行求解。基本调用如下：

fsolve(func, x0)

func(x)是计算方程组误差的函数，它的参数x是一个矢量，表示方程组的各个未知数的一组可能解，func返回将x代入方程组之后得到的误差；x0为未知数矢量的初始值。

求解一个简单的方程：$ \sin(x) - \cos(x) = 0.2$

```python
from scipy.optimize import fsolve
import numpy as np
def f(A):
    x = float(A[0])
    return [np.sin(x) - np.cos(x) - 0.2]

result = fsolve(f, [1])

print(result)
print(f(result))
```

再求解一个

$ 4x^2-2\sin y\cdot z=0 $

$ 5y+3=0 $

$ y\cdot z-1.5=0 $

```python
from scipy.optimize import fsolve
import numpy as np
def f(A):
    x = float(A[0])
    y = float(A[1])
    z = float(A[2])
    return [4*x*x - 2*np.sin(y*z), 5*y + 3, y*z - 1.5]

result = fsolve(f, [1, 1, 1])
print(result)
print(f(result))
```

### 实践二

模拟退火算法求最优解

```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['font.sans-serif'] = ['KaiTi']  # 设置字体为楷体
plt.rcParams['axes.unicode_minus'] = False
```

```python
def aimFunction(x):
    """假定某函数，假装不知道他的最优解"""
    y = x ** 4 - 4 * x ** 3 - 10 * x ** 2 + 7 * x + 10
    return y
```

```python
x = np.arange(-3, 6, 0.01)
y = aimFunction(x)

plt.figure('demo', facecolor='#eaeaea', figsize=(12, 8), dpi=300)
plt.plot(x, y, label='原图像')

T = 1000
T_min = 10
x = np.random.uniform(low=-100, high=100)
k = 50
y = 0
t = 0

while T >= T_min:
    for i in range(k):
        y = aimFunction(x)
        x_New = x + np.random.uniform(low=-0.055, high=0.055) * T

        if -100 <= x_New <= 100:
            y_New = aimFunction(x_New)
            if y_New - y < 0:
                x = x_New
            else:
                p = np.exp(-(y_New - y)/T)
                r = np.random.uniform(low=0, high=1)
                if r < p:
                    x = x_New

    t += 1
    T = 1000/(t + 1)

print(x, aimFunction(x))

plt.scatter(x, aimFunction(x), label="预测最优解")
plt.legend()
plt.show()
```

```python
# ===============================样本不平衡、多分类的情况========================
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# 创建不均衡样本
rng = np.random.RandomState(0)
n_samples_1 = 1000
n_samples_2 = 100
n_samples_3 = 100
X = np.r_[1.5 * rng.randn(n_samples_1, 2), 0.5 * rng.randn(n_samples_2, 2) + [2, 2],0.5 * rng.randn(n_samples_3, 2) + [-3, 3]]  # 三类样本点中心为(1.5,1.5)、(2,2)、(-3,3)
y = np.array([0] * (n_samples_1) + [1] * (n_samples_2)+ [2] * (n_samples_3))  # 前面的1000个为类别0，后面的100个为类别1，最后100个类别为2

# 创建模型获取分离超平面, 这里用的是线性模型, 显然不妥, 只为演示.
clf = svm.SVC(decision_function_shape='ovo',kernel='linear', C=1.0)  # decision_function_shape='ovo'为使用1对1多分类处理。会创建n(n-1)/2个二分类。ovr为一对所有的处理方式
clf.fit(X, y)

# 多分类的情况下，获取其中二分类器的个数。
dec = clf.decision_function([[1.5,1.5]])  # decision_function()的功能：计算样本点到分割超平面的函数距离。 包含几个2分类器，就有几个函数距离。
print('二分类器个数：',dec.shape[1])
print("------------------------------------------")

# 绘制，第一个二分类器的分割超平面
w = clf.coef_[0]
a = -w[0] / w[1]  # a可以理解为斜率. 直线方程是: w[0]*x[0]+w[1]*x[1]+clf.intercept_[0]=0
xx = np.linspace(-5, 5)
yy = a * xx - clf.intercept_[0] / w[1]  # 二维坐标下的直线方程

# 使用类权重，获取分割超平面
wclf = svm.SVC(kernel='linear', class_weight={1: 10})
wclf.fit(X, y)

# 绘制 分割分割超平面
ww = wclf.coef_[0]
wa = -ww[0] / ww[1]
wyy = wa * xx - wclf.intercept_[0] / ww[1]  # 带权重的直线

# 绘制第一个二分类器的分割超平面和样本点
plt.figure(figsize=(12, 10), dpi=300)
h0 = plt.plot(xx, yy, 'k-', label='no weights')
h1 = plt.plot(xx, wyy, 'k--', label='with weights')
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.legend()

plt.show()
```
