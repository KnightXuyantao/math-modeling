```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(25,40,16)
x1=np.linspace(25,40,1000)#便于绘制曲线
x2=np.linspace(25,60,1000)
y=np.array([8.00,7.78,7.61,7.42,7.20,6.98,6.76,6.55,6.39,6.01,5.95,5.75,5.54,5.36,5.28,4.79])#导入数据
```

```python
args=np.polyfit(x,y,1)#拟合曲线
args
```

```python
yhat=np.polyval(args,x)#求解函数值
```

```python
plt.plot(x,yhat,'r-',x,y,'o')
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
plt.legend(['1','2'])
plt.show()#绘制图像
```

```python
def f(x):
    return x*(args[0]*x+args[1])
y1=f(x1)
s=np.ones(1000)*200
plt.plot(x1,y1,'b-',x1,s,'--')#绘制图像
plt.show()
```

```python
#求解最大值
from scipy.optimize import fsolve
f1=lambda x:x*(args[0]*x+args[1])-200
print('交点x坐标： ',fsolve(f1,[38]))
```

```python
def f2(x):
    return x*(args[0]*x+args[1])-0.16*x**2-(-100)-0.2*x-50
y2=f2(x1)
plt.plot(x1,y1,'b-',x1,s,'--',x1,y2,'-*')#绘制图像
plt.show()
```
